```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,2])
plt.ylabel('numbers')
plt.xlabel('indices')
plt.title('Plot')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,8,27,64])
plt.ylabel('cube')
plt.xlabel('number')
plt.grid()
plt.title('Plot')
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0.0, 5, 0.2)
plt.plot(t,t,'b--', label='y=x')
plt.plot(t,t**2,'rs', label='y=x^2')
plt.plot(t,t**3,'g^', label='y=x^3')
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth=2.0)
plt.show()
```

```python
x1 = [1,2,3]
y1 = [1,4,9]
x2 = [1,2,3]
y2 = [1,8,27]
l = plt.plot(x1,y1,x2,y2)
plt.setp(l[0], color = 'r', linewidth= 2.0)
plt.setp(l[1], color = 'g', linewidth= 2.0)
plt.grid()
```
