
### HashMap


```python
l1 = list(range(100))
random.shuffle(l1)
l2 = list(range(50))
random.shuffle(l2)

HashMap = {}
for i in l2:
    HashMap[i] = 1

count = 0
for i in l1:
    if HashMap.get(i) != None:
        print(i)
        count += 1
print(count)
```
